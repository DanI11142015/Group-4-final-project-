```python
# Importing libraries
import numpy as np
import pandas as pd
from scipy.stats import mode
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

%matplotlib inline
```

```python
#reading in our training data 

path = "/Users/danielleb./Documents/GitHub/Group-4-final-project-/Updated Material/Data sets/Training.csv"

data = pd.read_csv(path).dropna(axis = 1)

#balancing our data set 

disease_amount = data["prognosis"].value_counts()
temp_df = pd.DataFrame({
	"Disease": disease_amount.index,
	"Counts": disease_amount.values
})

plt.figure(figsize = (14,6))
temp_df.plot.bar(x = "Disease", y = "Counts")
plt.xticks(rotation=90)
plt.show()
```

```python
#convert labels into numerical format 

encoder = LabelEncoder()
data["prognosis"] = encoder.fit_transform(data["prognosis"])
```

```python
#splitting the data to train our model 
X = data.iloc[:,:-1]
y = data.iloc[:, -1]
X_train, X_test, y_train, y_test =train_test_split(
X, y, test_size = 0.2, random_state = 24)

print(f"Train: {X_train.shape}, {y_train.shape}")
print(f"Test: {X_test.shape}, {y_test.shape}")
```
